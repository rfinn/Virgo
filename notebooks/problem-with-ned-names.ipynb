```python
# look up names for the galaxies where we used NSA v1 IDs as NED input

rematch_nedname_nsa = np.zeros_like(v.main['NEDname'])
count = 0
NSAasinput = np.zeros(len(v.main),'bool')
NSAra = np.zeros(len(v.main),'d')
NSAdec = np.zeros(len(v.main),'d')

for i in mismatch_indices:
    if count%100 == 0:
        print('working on ',count)

    if v.ned['NEDinput'][i].startswith('NSA'):
        galname = 'NSA'+str(v.main['NSAIDV0'][i])
        print('querying NED for ',galname)
        NSAasinput[i] = True
        result_table = Ned.query_object(galname)
        rematch_nedname_nsa[i] = result_table['Object Name'].data[0]
        NSAra[i] = result_table['RA'].data[0]
        NSAdec[i] = result_table['DEC'].data[0]
    count += 1
```
