```python
def compare_column(table,var,title='Comparison'):
    plt.figure(figsize=(7,5))

    x = table[var+'_1']
    xerr = table[var+'_ERR_1']
    y = table[var+'_2']
    yerr = table[var+'_ERR_2']
    plt.errorbar(x,y,yerr=yerr,xerr=xerr,fmt='bo')
    x1,x2 = plt.xlim()
    plt.xticks(fontsize=14)
    xl = np.linspace(x1,x2,20)
    plt.plot(xl,xl,'k--')
    plt.xlabel(var+'_1',fontsize=20)
    plt.ylabel(var+'_2',fontsize=20)
    plt.title(title,fontsize=20)
    plt.savefig(var+'comparison.png')
```

```python
tab = cv
compare_column(tab,'R24',title='Isophotal radius at 24mag/sqarcsec')
compare_column(tab,'F_R24',title='R-band flux w/in R24')
compare_column(tab,'HR17',title='Halpha isophotal rad @ log(flux) = -17')
compare_column(tab,'HF_R24',title='Halpha flux w/in R24')
compare_column(tab,'C30',title='Concentration C30')
```

```python
cn = join(ctable,ntable,keys='VFID')
```

```python
tab = cn
compare_column(tab,'R24',title='Isophotal radius at 24mag/sqarcsec')
compare_column(tab,'F_R24',title='R-band flux w/in R24')
compare_column(tab,'HR17',title='Halpha isophotal rad @ log(flux) = -17')
compare_column(tab,'HF_R24',title='Halpha flux w/in R24')
compare_column(tab,'C30',title='Concentration C30')
```
